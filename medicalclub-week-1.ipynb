## 1. Loading the Dataset Package

We use **MedMNIST**, a lightweight collection of biomedical image datasets designed for machine learning practice.  
For this workshop, we use **PneumoniaMNIST**, which contains small chest X-ray images labeled as:

- `0` = Normal
- `1` = Pneumonia

```python
# Install the MedMNIST package.
# In Colab, commands that start with ! are terminal commands, not normal Python code.
!pip install medmnist

# matplotlib is used to display images and graphs.
import matplotlib.pyplot as plt

# NumPy is used to convert image data into arrays that Python can process.
import numpy as np

# PneumoniaMNIST is the specific chest X-ray dataset used in this workshop.
from medmnist import PneumoniaMNIST
```

## 2. Downloading the Training Dataset

The **training set** is the part of the dataset used to teach the AI model.  
In Week 1, we are not training yet; we are only loading and exploring the data.

```python
# split='train' means we are loading the training portion of PneumoniaMNIST.
# download=True allows Colab to automatically download the dataset if it is not already available.
train_dataset = PneumoniaMNIST(split='train', download=True)
```

## 3. Checking the Number of Images

It is important to inspect the size of a dataset before building a model.  
This helps us understand how much data the model will learn from.

```python
# len(train_dataset) returns the number of images in the training set.
print("Number of training images:", len(train_dataset))
```

## 4. Looking at One Example

Each dataset item contains two parts:

1. `image`: the chest X-ray image  
2. `label`: the correct answer for that image

The image is converted into a NumPy array so we can inspect its shape and display it.

```python
# Get the first image and its label from the dataset.
image, label = train_dataset[0]

# Convert the image into a NumPy array.
# This makes it easier to check the image dimensions and use it with AI models later.
image_np = np.array(image)

# The shape tells us the image height, width, and sometimes color channels.
print("Image shape:", image_np.shape)

# The label tells us whether the image is Normal (0) or Pneumonia (1).
print("Label:", label)
```

## 5. Displaying One Image

Medical images such as chest X-rays are usually shown in grayscale.  
The `squeeze()` function removes unnecessary dimensions so the image can be displayed correctly.

```python
# Display the selected X-ray image.
plt.imshow(image_np.squeeze(), cmap='gray')

# Show the numeric label in the title.
plt.title(f"Label: {label[0]}")

# Hide axis numbers because they are not needed for image interpretation here.
plt.axis('off')

# Show the final image.
plt.show()
```

## 6. Displaying a Normal Image

Now we search through the dataset until we find an image labeled `0`, which means **Normal**.  
This helps students connect the numeric label with a visible X-ray example.

```python
# Create empty variables to store the first normal image we find.
normal_image = None
normal_label = None

# Go through the dataset one image at a time.
for i in range(len(train_dataset)):
    image, label = train_dataset[i]

    # Label 0 means the chest X-ray is normal.
    if label[0] == 0:
        normal_image = np.array(image)
        normal_label = label
        break  # Stop after finding the first normal image.

# If a normal image was found, display it.
if normal_image is not None:
    print(f"Found a normal image with label: {normal_label}")

    plt.imshow(normal_image.squeeze(), cmap='gray')
    plt.title(f"Label: {normal_label[0]} (Normal)")
    plt.axis('off')
    plt.show()
else:
    print("Could not find a normal image (label 0) in the training dataset.")
```

## 7. Displaying a Pneumonia Image

We repeat the same process for label `1`, which means **Pneumonia**.  
Comparing normal and pneumonia examples helps students understand what the model will later try to classify.

```python
# Create empty variables to store the first pneumonia image we find.
pneumonia_image = None
pneumonia_label = None

# Search through the dataset for an image labeled 1.
for i in range(len(train_dataset)):
    image, label = train_dataset[i]

    # Label 1 means the chest X-ray is labeled as pneumonia.
    if label[0] == 1:
        pneumonia_image = np.array(image)
        pneumonia_label = label
        break  # Stop after finding the first pneumonia image.

# If a pneumonia image was found, display it.
if pneumonia_image is not None:
    print(f"Found a pneumonia image with label: {pneumonia_label}")

    plt.imshow(pneumonia_image.squeeze(), cmap='gray')
    plt.title(f"Label: {pneumonia_label[0]} (Pneumonia)")
    plt.axis('off')
    plt.show()
else:
    print("Could not find a pneumonia image (label 1) in the training dataset.")
```
