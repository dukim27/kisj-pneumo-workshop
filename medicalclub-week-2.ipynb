## 4. Import TensorFlow / Keras

TensorFlow is the deep learning library used in this workshop.  
Keras provides beginner-friendly tools for stacking neural network layers.

```python
# Import TensorFlow, the main deep learning library.
import tensorflow as tf

# Import Keras layers and model tools for building a neural network.
from tensorflow.keras import layers, models
```

## 5. Build a Simple CNN Model

The model is built with layers that process the image step by step:

1. **Input layer:** receives a 28×28 grayscale X-ray image  
2. **Conv2D layers:** detect visual patterns  
3. **MaxPooling layers:** reduce image size while keeping strong signals  
4. **Flatten layer:** converts 2D feature maps into a 1D vector  
5. **Dense layers:** combine learned features and make the final prediction  
6. **Sigmoid output:** returns one probability between 0 and 1

```python
# Create a CNN model by stacking layers in order.
model = models.Sequential([
    # The input image is 28 pixels by 28 pixels with 1 grayscale channel.
    layers.Input(shape=(28, 28, 1)),

    # First convolutional layer:
    # 16 filters scan small 3x3 areas of the image to detect simple patterns.
    layers.Conv2D(16, (3, 3), activation='relu'),

    # MaxPooling reduces the feature map size and keeps the strongest signals.
    layers.MaxPooling2D((2, 2)),

    # Second convolutional layer:
    # 32 filters can learn more complex combinations of patterns.
    layers.Conv2D(32, (3, 3), activation='relu'),

    # Downsample again to reduce computation.
    layers.MaxPooling2D((2, 2)),

    # Flatten changes the 2D feature maps into a 1D vector.
    layers.Flatten(),

    # Dense layer combines all extracted features.
    layers.Dense(64, activation='relu'),

    # Output layer:
    # sigmoid returns a probability for binary classification.
    # Close to 0 = Normal, close to 1 = Pneumonia.
    layers.Dense(1, activation='sigmoid')
])
```

## 6. Show Model Structure

`model.summary()` shows each layer, its output shape, and the number of trainable parameters.  
This is useful for checking whether the model was built correctly.

```python
# Display the model architecture, output shapes, and number of parameters in each layer.
model.summary()
```

## 7. Compile the Model

Before training, the model must be compiled.

- **Optimizer:** updates the model weights
- **Loss:** measures how wrong the prediction is
- **Metrics:** show performance during training

```python
model.compile(
    optimizer='adam',            # Adam is a commonly used optimizer for neural networks.
    loss='binary_crossentropy',  # Binary cross-entropy is used for two-class classification.
    metrics=['accuracy']         # Accuracy shows how often the model predicts correctly.
)
```

## 8. Practice: Change the Model Size

This practice cell lets students experiment with model design.  
Changing the number of filters or Dense units changes the model capacity.

Public reader note: A larger model is not always better. It may learn more patterns, but it can also overfit or require more training time.

```python
# Practice model with larger layer sizes.
# Students can compare this model with the earlier version using model.summary().
practice_model = tf.keras.Sequential([
    # Input: 28x28 grayscale image.
    layers.Input(shape=(28, 28, 1)),

    # Increased from 16 filters to 32 filters.
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # Increased from 32 filters to 64 filters.
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),

    # Increased from 64 units to 128 units.
    layers.Dense(128, activation='relu'),

    # Binary output: Normal or Pneumonia.
    layers.Dense(1, activation='sigmoid')
])

practice_model.summary()
```
